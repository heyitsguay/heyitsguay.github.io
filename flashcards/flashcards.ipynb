```python
import json
import random
import time

import ipywidgets as ipw
```

```python
with open('wordslearned.json', 'r') as f:
    translation_pairs = json.load(f)
```

```python
pair_idx = 0
n_pairs = len(translation_pairs)

def clicked_next(button):
    global pair_idx
    pair_idx = (pair_idx + 1) % n_pairs
    center.value = f'<center>{translation_pairs[pair_idx][0]}</center>'
    pass

def clicked_prev(button):
    global pair_idx
    pair_idx = (pair_idx - 1 + n_pairs) % n_pairs
    center.value = f'<center>{translation_pairs[pair_idx][0]}</center>'
    pass

next_button = ipw.Button(
    description='Next',
    layout=ipw.Layout(width='80%', height='30%'))
next_button.on_click(clicked_next)
prev_button = ipw.Button(
    description='Previous',
    layout=ipw.Layout(width='80%', height='30%'))
prev_button.on_click(clicked_prev)
header = ipw.HTML('<h1>Flashcard app</h1>')
footer = ipw.Text(value='', placeholder='Enter text')
center = ipw.HTML(
    value=f'<center>{translation_pairs[pair_idx][0]}</center>')



ipw.AppLayout(header=header,
              left_sidebar=prev_button,
              center=center,
              right_sidebar=next_button,
              footer=footer,
              grid_gap='10px',
              justify_items='center',
              align_items='center')
```
